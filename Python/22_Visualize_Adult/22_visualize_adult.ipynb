```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('adult.csv', header=None)
df
```

```python
df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num' , 'marital-status', 'occupation', 'relationship', 'race', 'sex','capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income-category']
df
```

```python

plt.hist(df['age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
df['income-category'].value_counts().plot.pie(autopct='%1.1f%%', legend=True)
plt.title('Income Categories')
plt.show()
```

```python
# Calculate the proportions
proportions = df['income-category'].value_counts()

# Plot a pie chart
plt.pie(proportions, labels=proportions.index, autopct='%1.1f%%')
plt.title('Income Categories Distribution')
plt.show()
```

```python
plt.scatter(df['age'], df['hours-per-week'])
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.title('Scatter Plot of Age vs Hours per Week')
plt.show()
```

```python
sns.scatterplot(data=df, x='education-num', y='capital-loss', hue='income-category')
plt.show()
```

```python
sns.boxplot(data=df, x='hours-per-week', y='education-num')
plt.plot()
```

```python
# Python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot
sns.scatterplot(x='age', y='income-category', data=df)

# Add boxplots
sns.boxplot(x='age', y='income-category', data=df, whis=np.inf, color=".8")

plt.title('Age vs Income with Boxplots')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
sns.barplot(data=df, x='education-num', y='capital-loss')
plt.show()
```

```python
col = ['education-num', 'age']
for i in col:
    df[i].value_counts().plot(kind='bar')
    plt.show()
```
