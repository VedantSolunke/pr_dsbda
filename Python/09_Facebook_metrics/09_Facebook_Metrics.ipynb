```python
# Sample sales data
data = {
    'Date': ['2024-05-01', '2024-05-01', '2024-05-02', '2024-05-02', '2024-05-01'],
    'Product': ['A', 'B', 'A', 'B', 'A'],
    'Sales': [100, 150, 200, 120, 180],
    'Region': ['East', 'West', 'East', 'West', 'East']
}

df01 = pd.DataFrame(data)
print("Original DataFrame:")
df01
```

```python
# Create pivot table
pivot_table01 = pd.pivot_table(df01, index=['Product'], columns=['Region'], values='Sales', aggfunc='sum', fill_value=0)

print("\nPivot Table:")
pivot_table01
```
