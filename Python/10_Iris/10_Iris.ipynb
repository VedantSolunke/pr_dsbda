```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv', header=None)
df
```

```python
df.columns = ['sepalLength','sepalWidth', 'petalLength', 'petalWidth', 'class']
df
```

```python
df.isnull().sum()
```

```python
df['class'].unique()
```

## g. Create data subsets for different species

```python
df_sub01 = df[df['class'] == 'Iris-setosa']
df_sub01.head()
```

```python
df_sub02 = df[df['class'] == 'Iris-versicolor']
df_sub02.head()
```

```python
df_sub03 = df[df['class'] == 'Iris-virginica']
df_sub03.head()
```

## h. Merge two subsets

```python
df_merge = pd.concat([df_sub02, df_sub03])
df_merge
```

## i. Sort Data Petal Length

```python
df_sort = df.sort_values(by=['petalLength'])
df_sort
```

## j. Transposing Data

```python
df_transpose = df.transpose()
df_transpose
```

## k. Melting Data to long format

```python
df_melt = pd.melt(df, id_vars=['sepalLength', 'sepalWidth'])
df_melt
```

```python
df_melt02 = pd.melt(df, id_vars=['class'])
df_melt02
```

## l. Casting data to wide format

```python
df_long = pd.pivot_table(df, index='class')
df_long
```
