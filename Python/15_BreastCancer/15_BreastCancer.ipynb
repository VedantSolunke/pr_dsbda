```python
import pandas as pd
import numpy as np  
```

```python
df = pd.read_csv('BreastCancerWc.csv', header=None)
df
```

```python
df.columns = ['Sample code Number', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli', 'Mitoses', 'Class']
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
(df.values == np.NaN).sum()
```

```python
(df.values == '?' ).sum()
```

```python
df = df.replace('?', np.NaN)
```

```python
df.dropna(inplace=True)
```

```python
(df.values == '?').sum()
```

```python
df.shape
```

```python
df = df.astype(float)
```

```python
def zscore_filter(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    out = z_scores > threshold
    data = data[~out.any(axis=1)]
    return data

df = zscore_filter(df)
```

```python
df
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df['Class'].unique()
```

```python
df.loc[df['Class'] == 2, 'Class'] = 'Y'
df.loc[df['Class'] ==4, 'Class'] = 'N'
```

```python
df
```

```python
le = LabelEncoder()
le.fit(df['Class'])
le.classes_
```

```python
df['Class']  = le.transform(df['Class'])
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
# Step 1: Split the dataset into features (X) and target variable (y)

X = df.drop('Class', axis=1)
Y = df['Class']
```

```python
# Step 2: Split the dataset into training and testing sets
X_train, X_test, Y_train , Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# Step 3: Train a regression model (Logistic Regression) and evaluate its accuracy

regression_model = LogisticRegression()
regression_model.fit(X_train, Y_train)
```

```python
regression_predictions = regression_model.predict(X_test)
regression_accuracy = accuracy_score(Y_test, regression_predictions)
```

```python
# Step 4: Train a NaÃ¯ve Bayes model (Gaussian NaÃ¯ve Bayes) and evaluate its accuracy

naive_bayes_model = GaussianNB()
naive_bayes_model.fit(X_train, Y_train)
```

```python
naive_bayes_predictions = naive_bayes_model.predict(X_test)
naive_bayes_accuracy = accuracy_score(Y_test, naive_bayes_predictions)
```

```python

print("Regression Model Accuracy: ", regression_accuracy)
print("NaÃ¯ve Bayes Model Accuracy: ", naive_bayes_accuracy)
```
