```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('hepatitis.csv', header=None)
df
```

```python

df.columns = ['class', 'age', 'sex', 'steroid','antivirals', 'fatigue', 'malaise', 'anorexia','liver_big','liver_firm','spleen_palpable', 'spiders','ascites','varices','bilirubin','alk_phosphate', 'sgot','albumin','protime','histology']
df
```

```python
df.shape
```

## q. Data cleaning(Remove NA, ?, Negative values etc.)

```python
df.isnull().sum()
```

```python
# df[df == '?'].sum()
df = df.replace('?', np.nan)
df
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df
```

## r. Error correcting(Outlier detection and removal)

```python
df = df.astype(float)
df
```

```python
def remove_outlier(data):
    z_score = np.abs((data - data.mean())/data.std())
    out = z_score <= 3
    data = data[out.all(axis=1)]
    return data

df = remove_outlier(df)
```

```python
df
```

## s. Data transformation

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(df['class'])
le.classes_
```

```python
df['class'] = le.transform(df['class'])
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x = df.drop('class', axis=1)
y = df['class']
```

```python
x
```

```python
y
```

```python
x_train, x_text, y_train, y_test = train_test_split(x, y,  random_state=42)
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
```

```python
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
```

```python
y_pred_lr = lr_model.predict(x_text)
y_pred_nb = nb_model.predict(x_text)
```

```python
accuracy_lr = accuracy_score(y_test, y_pred_lr)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
```
